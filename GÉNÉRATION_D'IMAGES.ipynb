```python
# Importations
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# Vérification GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")

# Chargement du modèle
model_id = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    safety_checker=None,  # Désactive le filtre de sécurité pour plus de créativité
)
pipe = pipe.to(device)

# Optimisation pour GPU
if device == "cuda":
    pipe.enable_attention_slicing()
    pipe.enable_xformers_memory_efficient_attention()
```

```python
# Prompt de génération
prompt = "A beautiful cyberpunk city with neon lights, rain, 4K, detailed"

# Génération
image = pipe(
    prompt,
    height=512,
    width=512,
    num_inference_steps=50,
    guidance_scale=7.5,
    generator=torch.Generator(device=device).manual_seed(42)  # Seed pour reproductibilité
).images[0]

# Affichage
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')
plt.show()

# Sauvegarde
image.save("cyberpunk_city.png")
```
